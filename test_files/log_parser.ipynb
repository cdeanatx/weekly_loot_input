```python
import re
import pandas as pd
```

```python
loot_lines = pd.read_csv('loot_logs/10.12.txt', sep='\n', header=None)
loot_lines.columns = ['full_lines']
loot_lines = loot_lines.loc[loot_lines['full_lines'].str.match(r'^-.*')]
loot_lines
```

```python
loot_lines[['item','player']] = loot_lines['full_lines'].str.extract(r'^-\s(.+)\s-\s0\sDKP\s-\s(\w+)')

loot_lines = loot_lines[['item', 'player']]
```

```python
bad_items = ['Staff of Disintegration', 'Warp Slicer', 'Cosmic Infuser',
            'Phaseshift Bulwark', 'Devastation', 'Infinity Blade',
            'Netherstrand Longbow', 'Nether Spike', 'Staff of Disintegration',
            'Pit Lord\'s Satchel', 'Pattern: Belt of Deep Shadow',
            'Pattern: Belt of the Black Eagle']

loot_lines = loot_lines.loc[~loot_lines['item'].isin(bad_items)]
loot_lines = loot_lines.loc[loot_lines['player'] != 'Roll']
loot_lines
```
